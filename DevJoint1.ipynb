# Outlier Detection

```python
def iqr_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] < lower) | (df[col] > upper)]

total_price_outliers = iqr_outliers("total_price")
print("Total price outliers:", len(total_price_outliers))
```

```python
df['Total Price_log'] = np.log1p(df['total_price'])
```

```python
print('Total Price_log: ',len(iqr_outliers('Total Price_log'))) #since there can be luxurious properties, all of the outliers were not removed, instead extreme values are scaled
```

```python
df.drop(columns =['total_price'], inplace=True)
```

```python
def haversine_km(lat1, lng1, lat2, lng2):
    lat1, lng1, lat2, lng2 = map(np.radians, [lat1, lng1, lat2, lng2]) #np.sin and np.cos works with radians.
    distance_lat = lat2 - lat1
    distance_lng = lng2 - lng1
    a = np.sin(distance_lat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * (np.sin(distance_lng / 2) ** 2) #haversin formula used for finding distance between 2 locations
    return 2 * 6371 * np.arcsin(np.sqrt(a))  # Earth radius = 6371 km

BAKU_CENTER_LAT, BAKU_CENTER_LNG = 40.3777, 49.8920  # Fountain Square
df['distance_to_center_km'] = haversine_km(df['lat'], df['lng'], BAKU_CENTER_LAT, BAKU_CENTER_LNG).round(2)
df.head()
```
